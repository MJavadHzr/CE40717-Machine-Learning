```python
# Show statistical info for numeric and categorical columns

#dropping unused columns
data_set.drop(columns=['Unnamed: 0', 'card'], inplace=True, errors='ignore')

#determine columns type
categorical = list(data_set.columns[data_set.dtypes == object])
numeric = list(data_set.columns[data_set.dtypes != object])

#calculate stat
numeric_stat = data_set[numeric].agg(func=['min', 'max', 'mean', 'var', 'median'])
categ_stat = data_set.agg({'owner':'value_counts', 'selfemp':'value_counts'})

#output
print('numerical columns statistic:\n{}'.format(numeric_stat))
print('\ncategorical columns statistic:\n{}'.format(categ_stat))
```

```python
# Calc missing fields count for each col

def na_percentage(s):
    try:
        na_per = s.value_counts(normalize=True)[True]
        return na_per*100
    except:
        return 0

data_set.isna().agg(func=na_percentage)
```

```python
#-------[!!!!]------: do not run this box after you have run the next box, cause it will change quartiles
outlier_col = numeric.copy()
outlier_col.remove('reports')
outlier_col.remove('majorcards')

q1 = np.nanpercentile(data_set[outlier_col], axis=0, q=25).reshape(1,-1)
q3 = np.nanpercentile(data_set[outlier_col], axis=0, q=75).reshape(1,-1)
iqr = q3 - q1

is_outlier = (data_set[outlier_col].values < np.fmax(0, q1-1.5*iqr)) | (q3+3*iqr < data_set[outlier_col].values)
```

<div dir="rtl">
در برخی از ستون‌های اسمی عباراتی وجود دارند که معادل ناموجود بودن مقدار آن ردیف در ان ستون هستند. همچنین ستون‌های اسمی باینری نیز بهتر است به حالت True, False در بیایند.
        برای ستون‌های عددی نیز برخی مقادیر غیر مجاز یا ناموجود هستند(یعنی برای ستون‌هایی که باید مقدار مثبت داشته باشند، مقدار منفی ثبت شده است و یا داده‌ای بسیار بزرگ به عبارتی outlier وجود دارد) که باید به صورت None در بیایند.
    این ستون‌ها را تغییر دهید و دوباره نسبت مقادیر ناموجود هر ستون و همینطور اطلاعات آماری ستون های عددی تغییر یافته را محاسبه کنید.
    <br>
    (۵ نمره)
</div>

```python
# Calc missing fields

#------[categorical]-------
data_set['owner'].replace(['yes','1','0','no'], [True,True,False,False], inplace=True)
data_set['selfemp'].replace(['yes','no'], [True,False], inplace=True)

data_set.loc[~data_set['owner'].isin([True,False]), 'owner'] = np.nan
data_set.loc[~data_set['selfemp'].isin([True,False]), 'selfemp'] = np.nan

#------[numeric]-------
# handle negative values
data_set.mask(data_set < 0, np.nan, inplace=True)

# handle strong outliers
for i in range(0, len(outlier_col)):
    data_set.loc[is_outlier[:,i], outlier_col[i]] = np.nan


#-----[check nans]-----
data_set.isna().agg(func=na_percentage)
```

```python
# Show numeric cols' statistical info

numeric_stat = data_set[numeric].agg(func=['min', 'max', 'mean', 'var', 'median'])
print('numerical columns statistic:\n{}'.format(numeric_stat))
```

```python
# Load and show some of data points
from PIL import Image
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
with np.load('mnist.zip') as data:
    arr, label = data['X'], data['y']

Image.fromarray(arr[0])
```

<div dir=rtl>
با دو روش pca و tsne و با استفاده از ۲ تا از مقادیر ویژه ابعاد دیتاست را کاهش دهید.
    برای روش pca مقادیر و بردارهای ویژه‌ی استفاده شده و درصد واریانس حفظ شده را مشخص کنید.
    <br>
    (۵ نمره)
</div>

```python
from sklearn.preprocessing import StandardScaler
```

```python
data_count = arr.shape[0]
features_count = arr.shape[1]*arr.shape[2]
x = np.reshape(arr, (data_count, features_count))
x = StandardScaler().fit_transform(x)
y = np.reshape(label, (data_count, 1))

#-----[PCA]-----
pca_util = PCA(n_components=2)
pca_result = pca_util.fit_transform(x)
pca_data = pd.DataFrame(data=np.hstack((pca_result, y)), columns=['pc1', 'pc2', 'target'])

#-----[tSNE]-----
tsne_util = TSNE(n_components=2, n_iter=500)
tsne_result = tsne_util.fit_transform(x)
tsne_data = pd.DataFrame(data=np.hstack((tsne_result, y)), columns=['tsne1', 'tsne2', 'target'])
```

```python
print('selected eigen vectors:{}'.format(pca_util.components_))
print('variance ratio:{}'.format(pca_util.explained_variance_ratio_))
print('selected singular values:{}'.format(pca_util.singular_values_))
```

<div dir=rtl>
دیتاست‌های حاصل را در قالب نمودار دوبعدی نمایش داده و قابلیت این دو روش را در جداسازی کلاس‌های مختلف و با استفاده از لیبل داده‌ها مقایسه کنید. علت برتری یک روش به روش دیگر را به طور خلاصه توضیح دهید.
    <br>
    (۴ نمره)
</div>

```python
fig, axes = plt.subplots(1,2, figsize=(14,7))
sns.scatterplot(data=pca_data, x='pc1', y='pc2', hue='target',palette='Set2' , ax=axes[0]);
sns.scatterplot(data=tsne_data, x='tsne1', y='tsne2', hue='target',palette='Set2', ax=axes[1]);
axes[0].set_title('PCA');
axes[1].set_title('t-SNE');
```

### PCS vs t-SNE
<div dir=rtl>
    همانطور که در تصویر نیز دیده می‌شود، در روش
    t-SNE
    ساختار همسایگی داده‌ها حفظ شده و داده‌ها به دسته‌بندی‌های مجزایی کاهش پیدا کرده‌اند.
    اما در روش
    pca
    به این ساختار توجه نشده و این باعث شده است در فضای کاهش یافته، دسته‌های مختلف داده به صورت درهم توزیع شده و نتوان این دسته‌ها را به راحتی تفکیک کرد.
    بنابراین مشخص است که روش
    t-SNE
    بهتر عمل کرده است. اما باید توجه کرد که این روش برخلاف
    pca
    که کامل معین و فاقد هرگونه پارامتر تعیین‌کننده‌ای هست،
    دارای هایپر پارامتر‌هایی است که باید به صورت دستی مشخص شوند و یا چندیدن بار این کاهش را با پارامترهای مختلف انجام داد تا به بهترین نتیجه ممکن رسید.
</div>

<div dir=rtl>
حال میخواهیم pca را بر روی دیتاست اصلی خود بدون استفاده از کتابخانه‌ی sklearn پیاده سازی کنیم. برای اینکار در ابتدا ستون‌های فیچر را جداکرده و سپس داده‌ها را نرمالایز کنید. لزوم نرمالایز کردن داده‌ها برای روش pca به چه خصوصیتی از این روش باز میگردد؟
    <br>
    (۴ نمره)
</div>

```python
data_count = data_set.shape[0]
features = data_set.drop(columns=['income', 'wealth', 'age_category', 'exp_percentiles']).values

f_mean = features.sum(axis=0)/data_count
f_normal = (features - f_mean)/np.std(features, axis=0)
```

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
covv = np.array((f_normal.T @ f_normal)/data_count, dtype=float)
e_values, e_vectors = np.linalg.eig(covv)
sorted_indices = np.argsort(e_values)[::-1]
e_values = e_values[sorted_indices]
e_vectors = e_vectors[:,sorted_indices]

e_values, e_vectors
```

<div dir=rtl>
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
    <br>
    (۴ نمره)
</div>

```python
plt.pie(e_values, autopct='%1.1f%%');
```

<div dir=rtl>
با حفظ حداقل ۵۰ درصد واریانس داده‌ها، ماتریس پرتو(projection matrix) را بدست آورید و تبدیل pca را بر روی دیتاست اجرا کنید.
    <br>
    (۵ نمره)
</div>

```python
# we should keep the first 4 eigen values
projection_m = e_vectors[:,0:4]
pca_features = features @ projection_m
reduced_data = pd.DataFrame(data=pca_features, columns=['pc1', 'pc2', 'pc3', 'pc4'])
reduced_data
```

    <br>
    (۴ نمره)
</div>
